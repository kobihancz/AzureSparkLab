#### Start a simple Spark Session

```pyspark
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

#### Load the Walmart Stock CSV File, have Spark infer the data types.

```pyspark
df = spark.read.format('csv').option("inferSchema", True).option("header",True).option("sep",',').load("wasb://sparklab@sparklabhdinstorage.blob.core.windows.net/HdiNotebooks/walmart_stock.csv")
```

#### What are the column names?

```pyspark
df.columns
```

#### What does the Schema look like?

```pyspark
df.printSchema()
```

#### Print out the first 5 columns.

```pyspark
result = df.select('Date', 'Open', 'High', 'Low', 'Close')
result.show()
```

#### Use describe() to learn about the DataFrame.

```pyspark
describe_df = df.describe()
describe_df.printSchema()
describe_df.show()
```

#### Create a new dataframe with a column called HV Ratio that is the ratio of the High Price versus volume of stock traded for a day.

```pyspark
df2 = df.withColumn("HV Ratio", col("High")/col("Volume")).select("HV Ratio")
df2.show()
```

#### What day had the Peak High in Price?

```pyspark
df.orderBy(col('High').desc()).show(1)
```

#### What is the mean of the Close column?

```pyspark
df.select(avg("Close")).show()
```

#### What is the max and min of the Volume column?

```pyspark
df.select(max("Volume"),min("Volume")).show()
```

#### How many days was the Close lower than 60 dollars?

```pyspark
df.filter(col('Close') < 60).count()
```

#### What percentage of the time was the High greater than 80 dollars ?
#### In other words, (Number of Days High>80)/(Total Days in the dataset)

```pyspark
df.select(df.filter(col('High')>80).count()/count('Date')*100).show()
```

#### What is the Pearson correlation between High and Volume?
#### [Hint](http://spark.apache.org/docs/latest/api/python/pyspark.sql.html#pyspark.sql.DataFrameStatFunctions.corr)

```pyspark
df.stat.corr("High","Volume")
```

#### What is the max High per year?

```pyspark
df3=df.select(year(col('Date')).alias('Year'),col('High'))
df3.groupBy('Year').max('High').orderBy('Year').show()
```

#### What is the average Close for each Calendar Month?
#### In other words, across all the years, what is the average Close price for Jan,Feb, Mar, etc... Your result will have a value for each of these months. 

```pyspark
df4=df.select(month(col('Date')).alias('Month'),col('Close'))
df4.groupBy('Month').avg('Close').orderBy('Month').show()
```
